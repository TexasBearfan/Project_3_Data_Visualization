```python
import pandas as pd
```

```python
file='data/FOIA-FAQ-NPS-Mortality-Data-CY2007-to-CY2023-Released-August-2023.csv'
```

```python
df=pd.read_csv(file,header=2)
df.head()
```

```python
lake_mead=df.loc[df['Park Name']=='Lake Mead National Recreation Area']
lake_mead['Cause of Death'].value_counts()
```

```python
grouped_df=df.groupby(['Park Name']).count()
sorted_df=grouped_df.sort_values(by='Cause of Death',ascending=False)
sorted_df
```
